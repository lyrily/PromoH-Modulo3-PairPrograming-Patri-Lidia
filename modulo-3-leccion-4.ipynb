Ejercicios de Limpieza
Después de la unión de datos, tenemos dos columnas de "country". Elimina una de ellas.

```python
df_3 = df_concat1.drop(['country'], axis= 1)
df_3
```

Los nombres de las columnas no son homogeneos. Cambia los nombres de las columnas de tal forma que:

No tengan espacios.

Estén en minúscula.

No tengan paréntesis, es decir, quitar "(%)", "(Km2)".

Algunas columnas tiene "\n". Eliminalos de los nombres de las columnas.

Algunas columnas tienen ":". Eliminalos de los nombres de las columnas.

```python
df_3.columns
```

```python
nuevas_columnas = {columna: columna.lower().replace(".", "").replace('_x', '').replace('_y', '').replace(' ', '').replace('(Km2)', '').replace('/n', '').replace('( %)','').replace('(%)','').replace('\n(p/km2)','').replace('(km2)', '').replace(':','') for columna in df_3.columns}
nuevas_columnas
```

```python
df_3.rename(columns = nuevas_columnas, inplace = True)
df_3.head()
```

La columnas coordinates tiene la latitud y la longitud en una sola columna. Crea dos columnas nuevas, una con la longitud y otra con la latitud. Una vez hecho, elimina la columna de coordinates.

```python
df_3[["latitude","longitude"]]=df_3["coordinates"].str.split( "," ,expand=True)
```

```python
df_3
```

```python
df_3=df_3.drop(columns=["coordinates"])
```

```python
df_3
```

```python
#Las columnas unemployment_rate, total_tax_rate, tax_revenue, population_labor_force_participation, out_of_pocket_health_expenditure, gross_tertiary_education_enrollment
#gross_primary_education_enrollment, forested_area, cpi_change, agricultural_land tienen "%". 
#Elimina los "%" de los valores de las columnas.
columnas_porcentaje=["unemploymentrate", "totaltaxrate", "taxrevenue", "populationlaborforceparticipation", "outofpockethealthexpenditure", "grosstertiaryeducationenrollment",
"grossprimaryeducationenrollment", "forestedarea", "cpichange", "agriculturalland"]
for column in columnas_porcentaje:
    df_3[column]=df_3[column].str.replace("%","")
```

```python
df_3
```

```python
#Haz lo mismo para las columnas de gasoline_price, gdp, minimum_wage, pero eliminando "$".
columnas_dolar=["gasolineprice","gdp", "minimumwage"]
for column in columnas_dolar:
    df_3[column]=df_3[column].str.replace("$","")
```
